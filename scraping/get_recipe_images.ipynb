```python
import pandas as pd
from tqdm import tqdm
import requests
import cv2
```

```python
data = pd.read_csv('recipe_data_nutrition.csv')
```

```python
data = data.rename(columns={'Unnamed: 0': "id"})
```

```python
data.head()
```

```python
image_urls = data['img_link']
for idx, id in enumerate(tqdm(data['id'])):
    image_url = image_urls[idx]
    img_data = requests.get(image_url).content
    with open(f"Recipe_Images/{id}.jpg", 'wb') as handler:
        handler.write(img_data)
```

```python
data['img_path'] = './Recipe_Images/' + data['id'].astype('str') + '.jpg'
```

```python
data
```

```python
def resize_imgs(img_path_array, save_dir, img_size = (224, 224)):
    for img_path in tqdm(img_path_array):
        save_img_name = img_path[16:]
        save_path = f"{save_dir}/{save_img_name}"
        image = cv2.imread(img_path)
        image = cv2.resize(image, img_size)
        cv2.imwrite(save_path, image)
```

```python
resize_imgs(data['img_path'].values, './Recipe_Images_224_x_224')
```
